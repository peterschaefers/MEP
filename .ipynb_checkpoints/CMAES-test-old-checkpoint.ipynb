```python
def plotSingleFidelity(frequency: np.ndarray, amplitude: np.ndarray, fidelity: np.ndarray, title: str) -> None:
    """
    Plot a single heatmap using imshow for frequency, amplitude, and fidelity.
    """
    # Reshape fidelity to 2D (if not already a grid)
    Z = fidelity.reshape(int(np.sqrt(len(amplitude))), int(np.sqrt(len(frequency))))

    # Calculate the extent of the axes for imshow
    dx = (frequency.max() - frequency.min()) / (Z.shape[1] - 1)  # Step size along x-axis
    dy = (amplitude.max() - amplitude.min()) / (Z.shape[0] - 1)  # Step size along y-axis

    extent = [
        frequency.min() - 0.5 * dx, frequency.max() + 0.5 * dx,  # x-axis range
        amplitude.min() - 0.5 * dy, amplitude.max() + 0.5 * dy   # y-axis range
    ]

    # Plot with imshow
    plt.figure(figsize=(10, 8), dpi=300)
    cmap = 'viridis'
    vmin, vmax = 0.5, 1  # Set fidelity range for consistent scaling

    im = plt.imshow(Z, extent=extent, origin='lower', aspect='auto', cmap=cmap, vmin=vmin, vmax=vmax)
    plt.colorbar(im, label="Fidelity")
    plt.title(title, fontsize=16)
    plt.xlabel("Frequency [Hz]", fontsize=12)
    plt.ylabel("Amplitude [V]", fontsize=12)
    # filename = title.replace(" ", "_").lower() + ".pdf"
    # print("Saving plot as:", filename)
    # plt.savefig(filename)
    plt.show()
```

```python
title = "Average Fidelity; Duration:" + str(unique_duration_values[-1])
print(title)
plotSingleFidelity(freq_reshape[-1], amplitude_reshape[-1], average_fidelity_reshape[-1], title)
```

```python
def generateContinuousFunction(X0, y0):
    """
    Generate a continuous 2D function f(x1, x2) -> y based on measured data.

    Parameters:
        X0 (array-like): 2D array of shape (n_samples, 2), where each row is [frequency, amplitude].
        y0 (array-like): 1D array of shape (n_samples,), corresponding to fidelity values.

    Returns:
        callable: A continuous function f(x1, x2) -> y.
    """
    if len(X0) != len(y0):
        raise ValueError("X0 and y0 must have the same length.")
    X0 = np.asarray(X0)
    x1, x2 = X0[:, 0], X0[:, 1]  # Extract frequency and amplitude
    
    # Create the RBF interpolator
    rbf = Rbf(x1, x2, y0, function='multiquadric', smooth=0.1)
    
    def continuous_function(freq, amp):
        """
        Evaluate the continuous function at given frequency and amplitude.

        Parameters:
            freq (float or array-like): Frequency values.
            amp (float or array-like): Amplitude values.

        Returns:
            float or array-like: Interpolated fidelity values.
        """
        return rbf(freq, amp)
    
    return continuous_function
```

```python
def linear_interpolation_nearest_neighbors(X0, y0, point):
    """
    Estimate y_ab for a given point (a, b) using linear interpolation based on the two nearest neighbors.

    Parameters:
        X0 (array-like): 2D array of shape (n_samples, 2), where each row is [frequency, amplitude].
        y0 (array-like): 1D array of shape (n_samples,), corresponding to fidelity values.
        point (tuple): A tuple (a, b) representing the query point.

    Returns:
        float: The interpolated value y_ab at point (a, b).
    """
    point = np.asarray(point)
    
    # Calculate the Euclidean distance from the query point to all points in X0
    distances = np.linalg.norm(X0 - point, axis=1)
    
    # Find the indices of the two closest points
    nearest_indices = np.argsort(distances)[:2]
    p1, p2 = X0[nearest_indices]
    y1, y2 = y0[nearest_indices]
    
    # Perform linear interpolation based on the Euclidean distance
    dist1 = np.linalg.norm(p1 - point)
    dist2 = np.linalg.norm(p2 - point)
    
    # Linear interpolation formula (weighted average based on distance)
    y_ab = (y1 * dist2 + y2 * dist1) / (dist1 + dist2)
    
    return y_ab

X0 = np.column_stack((freq_reshape[-1],amplitude_reshape[-1]))
y0 = average_fidelity_reshape[-1]

point = (np.mean(freq_reshape[-1]),np.mean(amplitude_reshape[-1]))

linear_interpolation_nearest_neighbors(X0,y0,point)
```

```python
# def interpolateFidelity(frequencies, amplitudes, X0, y0, method='linear'):
#     points = np.column_stack((frequencies, amplitudes))
#     interpolated_values = []
#     for point in points:
#         # Calculate the Euclidean distance from the query point to all points in X0
#         distances = np.linalg.norm(X0 - point, axis=1)
        
#         # Find the indices of the four closest points
#         nearest_indices = np.argsort(distances)[:4]

#         #Find fidelity
#         interpolated_value = griddata(X0[nearest_indices],y0[nearest_indices],point,method=method)
#         interpolated_values.append(interpolated_value)
#     return interpolated_values

def interpolateFidelity(frequencies, amplitudes, X0, y0, method='linear'):
    # Combine frequencies and amplitudes into a 2D array (n_samples, 2)
    points = np.column_stack((frequencies, amplitudes))
    
    # Initialize an empty list to store the interpolated values
    interpolated_values = []
    
    # Loop through each query point
    for point in points:
        if any(np.allclose(point, X0[i]) for i in range(len(X0))):
            # If the point is in X0, use the corresponding fidelity value directly
            interpolated_value = y0[np.argmin(np.linalg.norm(X0 - point, axis=1))]
            interpolated_values.append(interpolated_value)
        else:
            # Calculate the Euclidean distance from the query point to all points in X0
            distances = np.linalg.norm(X0 - point, axis=1)
            
            # Find the indices of the four closest points
            nearest_indices = np.argsort(distances)[:4]
            
            # Interpolate the fidelity at the query point using the closest 4 points
            interpolated_value = griddata(X0[nearest_indices], y0[nearest_indices], point, method=method)
            
            # Append the interpolated value to the list
            interpolated_values.append(interpolated_value)
    
    # Return the interpolated values as a 1D numpy array
    return np.array(interpolated_values)

# import numpy as np
# from scipy.interpolate import interp1d

# def interpolateFidelity(frequencies, amplitudes, X0, y0, method='linear'):
#     # Combine frequencies and amplitudes into a 2D array (n_samples, 2)
#     points = np.column_stack((frequencies, amplitudes))
    
#     # Initialize an empty list to store the interpolated values
#     interpolated_values = []
    
#     # Loop through each query point
#     for point in points:
#         # Check if the point is exactly in X0
#         if any(np.allclose(point, X0[i]) for i in range(len(X0))):
#             # If the point is in X0, use the corresponding fidelity value directly
#             interpolated_value = y0[np.argmin(np.linalg.norm(X0 - point, axis=1))]
#             interpolated_values.append(interpolated_value)
#         else:
#             # Calculate the Euclidean distance from the query point to all points in X0
#             distances = np.linalg.norm(X0 - point, axis=1)
            
#             # Find the indices of the four closest points
#             nearest_indices = np.argsort(distances)[:4]
#             nearest_points = X0[nearest_indices]
#             nearest_fidelity = y0[nearest_indices]
            
#             # Check if all nearest points have the same frequency or the same amplitude
#             same_frequency = np.all(np.isclose(nearest_points[:, 0], nearest_points[0, 0]))  # Same frequency
#             same_amplitude = np.all(np.isclose(nearest_points[:, 1], nearest_points[0, 1]))  # Same amplitude

#             if same_frequency:
#                 # Perform 1D interpolation along the amplitude axis (second dimension)
#                 amplitude_values = nearest_points[:, 1]
#                 interp = interp1d(amplitude_values, nearest_fidelity, kind='linear', fill_value="extrapolate")
#                 interpolated_value = interp(point[1])  # Interpolate at the given amplitude
#             elif same_amplitude:
#                 # Perform 1D interpolation along the frequency axis (first dimension)
#                 frequency_values = nearest_points[:, 0]
#                 interp = interp1d(frequency_values, nearest_fidelity, kind='linear', fill_value="extrapolate")
#                 interpolated_value = interp(point[0])  # Interpolate at the given frequency
#             else:
#                 # Perform 2D interpolation using griddata if neither is the case
#                 interpolated_value = griddata(nearest_points, nearest_fidelity, point, method=method)
                
#             # Append the interpolated value to the list
#             interpolated_values.append(interpolated_value)
    
#     # Return the interpolated values as a 1D numpy array
#     return np.array(interpolated_values)
```
