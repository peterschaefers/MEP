```python
import networkx as nx
import matplotlib.pyplot as plt

# Define transitions as edges: (from_state, to_state, label)
transitions = [
    ('A', 'B', '0'),
    ('A', 'C', '1'),
    ('B', 'A', '0'),
    ('B', 'D', '1'),
    ('C', 'D', '0'),
    ('C', 'A', '1'),
    ('D', 'D', '0,1')  # self-loop
]

G = nx.MultiDiGraph()  # Allow multiple directed edges

for from_state, to_state, label in transitions:
    G.add_edge(from_state, to_state, label=label)

pos = nx.spring_layout(G)  # or nx.circular_layout(G), nx.shell_layout(G), etc.

# Draw nodes and edges
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=2000)
nx.draw_networkx_edges(G, pos, connectionstyle='arc3, rad=0.1', arrows=True)

# Draw node labels
nx.draw_networkx_labels(G, pos, font_size=12)

# Draw edge labels
edge_labels = {(u, v): d['label'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

plt.axis('off')
plt.title("Transition Diagram")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize=(6, 4))

# Positions
input_positions = {'|1⟩': (0.1, 0.75), '|0⟩': (0.1, 0.25)}
output_positions = {'|1⟩': (0.9, 0.75), '|0⟩': (0.9, 0.25)}

# Draw input and output labels
for state, pos in input_positions.items():
    ax.text(pos[0], pos[1], state, fontsize=14, ha='center', va='center', bbox=dict(boxstyle="round,pad=0.3", fc="lightblue"))

for state, pos in output_positions.items():
    ax.text(pos[0], pos[1], state, fontsize=14, ha='center', va='center', bbox=dict(boxstyle="round,pad=0.3", fc="lightgreen"))

# Draw arrows for transitions
def draw_arrow(start, end):
    ax.annotate("",
                xy=end, xycoords='data',
                xytext=start, textcoords='data',
                arrowprops=dict(arrowstyle="->", lw=2))

# Transitions
draw_arrow(input_positions['|0⟩'], output_positions['|0⟩'])  # |0⟩ → |0⟩
draw_arrow(input_positions['|1⟩'], output_positions['|0⟩'])  # |1⟩ → |0⟩
draw_arrow(input_positions['|1⟩'], output_positions['|1⟩'])  # |1⟩ → |1⟩

# Styling
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')
plt.title("Quantum State Transition Diagram", fontsize=14)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(6, 4))

# Box size adjustments
x_offset = 0.07  # distance from box center to edge for arrow start/end

# State positions (center of boxes)
states = ['|1⟩', '|0⟩']
y_coords = {'|1⟩': 0.75, '|0⟩': 0.25}
x_input = 0.2
x_output = 0.8

# Draw input and output boxes
for state in states:
    ax.text(x_input, y_coords[state], state, fontsize=14, ha='center', va='center',
            bbox=dict(boxstyle="round,pad=0.3", fc="lightblue", lw=1.5))
    ax.text(x_output, y_coords[state], state, fontsize=14, ha='center', va='center',
            bbox=dict(boxstyle="round,pad=0.3", fc="lightgreen", lw=1.5))

# Function to draw arrows between boxes with same vertical coords
def draw_arrow(y_start, y_end):
    start = (x_input + x_offset, y_start)
    end = (x_output - x_offset, y_end)
    ax.annotate("",
                xy=end, xycoords='data',
                xytext=start, textcoords='data',
                arrowprops=dict(arrowstyle="->", lw=2))

# Transitions
draw_arrow(y_coords['|0⟩'], y_coords['|0⟩'])  # |0⟩ → |0⟩
draw_arrow(y_coords['|1⟩'], y_coords['|0⟩'])  # |1⟩ → |0⟩
draw_arrow(y_coords['|1⟩'], y_coords['|1⟩'])  # |1⟩ → |1⟩

# Clean up plot
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(6, 4))

# Positions and offsets
x_input = 0.2
x_output = 0.8
x_offset = 0.07
arrow_head_separation = 0.025  # vertical offset to avoid arrowhead overlap

states = ['|1⟩', '|0⟩']
y_coords = {'|1⟩': 0.75, '|0⟩': 0.25}

# Draw input/output boxes
for state in states:
    ax.text(x_input, y_coords[state], state, fontsize=14, ha='center', va='center',
            bbox=dict(boxstyle="round,pad=0.3", fc="lightblue", lw=1.5))
    ax.text(x_output, y_coords[state], state, fontsize=14, ha='center', va='center',
            bbox=dict(boxstyle="round,pad=0.3", fc="lightgreen", lw=1.5))

# Arrow drawing with optional y-offset at endpoint
def draw_arrow(y_start, y_end, y_end_offset=0):
    start = (x_input + x_offset, y_start)
    end = (x_output - x_offset, y_end + y_end_offset)
    ax.annotate("",
                xy=end, xycoords='data',
                xytext=start, textcoords='data',
                arrowprops=dict(arrowstyle="->", lw=2))

# Transitions
draw_arrow(y_coords['|0⟩'], y_coords['|0⟩'], y_end_offset=0)                          # |0⟩ → |0⟩
draw_arrow(y_coords['|1⟩'], y_coords['|0⟩'], y_end_offset=arrow_head_separation)      # |1⟩ → |0⟩
draw_arrow(y_coords['|1⟩'], y_coords['|1⟩'])                                          # |1⟩ → |1⟩

# Final styling
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')
plt.tight_layout()
plt.show()
```
