```python
def fidelityEstimate3D(point, frequency, amplitude, duration, fidelity):
    # Get unique values of frequency, amplitude, and duration
    unique_freq = np.unique(frequency)
    unique_ampl = np.unique(amplitude)
    unique_dur = np.unique(duration)
    
    # Get the closest 3 unique frequencies, amplitudes, and durations
    dFreq = np.abs(unique_freq - point[0])
    dAmpl = np.abs(unique_ampl - point[1])
    dDur = np.abs(unique_dur - point[2])
    
    freqArgs = np.argsort(dFreq)[:2]
    amplArgs = np.argsort(dAmpl)[:2]
    durArgs = np.argsort(dDur)[:2]
    
    x_coor = unique_freq[freqArgs]
    y_coor = unique_ampl[amplArgs]
    z_coor = unique_dur[durArgs]
    
    # Sort the coordinates
    sortedX = np.sort(x_coor)
    sortedY = np.sort(y_coor)
    sortedZ = np.sort(z_coor)
    
    # Calculate differences
    dx = np.diff(sortedX)
    dy = np.diff(sortedY)
    dz = np.diff(sortedZ)
    
    fidelities = []
    
    # Iterate through all combinations of (frequency, amplitude, duration)
    for i in itertools.product(sortedX, sortedY, sortedZ):
        temp1 = np.where(i[0] == frequency)
        temp2 = np.where(i[1] == amplitude)
        temp3 = np.where(i[2] == duration)
        arg = np.intersect1d(np.intersect1d(temp1, temp2), temp3)
        fidelities.append(fidelity[arg])
    
    fidelities = np.array(fidelities)
    
    # Reshape fidelities into a 2x2x2 grid for interpolation
    fidelity_matrix = fidelities.reshape((2, 2, 2))
    
    # 3D Bilinear interpolation formula
    result = (fidelity_matrix[0, 0, 0] * (sortedX[1] - point[0]) * (sortedY[1] - point[1]) * (sortedZ[1] - point[2]) +
              fidelity_matrix[1, 0, 0] * (point[0] - sortedX[0]) * (sortedY[1] - point[1]) * (sortedZ[1] - point[2]) +
              fidelity_matrix[0, 1, 0] * (sortedX[1] - point[0]) * (point[1] - sortedY[0]) * (sortedZ[1] - point[2]) +
              fidelity_matrix[1, 1, 0] * (point[0] - sortedX[0]) * (point[1] - sortedY[0]) * (sortedZ[1] - point[2]) +
              fidelity_matrix[0, 0, 1] * (sortedX[1] - point[0]) * (sortedY[1] - point[1]) * (point[2] - sortedZ[0]) +
              fidelity_matrix[1, 0, 1] * (point[0] - sortedX[0]) * (sortedY[1] - point[1]) * (point[2] - sortedZ[0]) +
              fidelity_matrix[0, 1, 1] * (sortedX[1] - point[0]) * (point[1] - sortedY[0]) * (point[2] - sortedZ[0]) +
              fidelity_matrix[1, 1, 1] * (point[0] - sortedX[0]) * (point[1] - sortedY[0]) * (point[2] - sortedZ[0])) / (dx * dy * dz)
    
    return result
```

```python
def objective_2qb(points, frequency1, amplitude1, duration1, fidelity1, frequency2, amplitude2, duration2, fidelity2):
    objective = np.zeros(len(points))
    for i in range(len(points)):
        F1 = fidelityEstimate3D(points[i][:3], frequency1, amplitude1, duration1, fidelity1)
        F2 = fidelityEstimate3D(points[i][-3:], frequency2, amplitude2, duration2, fidelity2) 
        objective[i] = np.log(1-F1)[0] + np.log(1-F2)[0]
    return objective
    
def cmaes_frequency_amplitude_and_duration_2qb(mean, sigma, popsize, Ngen, 
                                           meas_freq1, meas_ampl1, meas_dur1, meas_fid1, 
                                           meas_freq2, meas_ampl2, meas_dur2, meas_fid2):
    bound = [0,1]
    
    minFreq1 = np.min(meas_freq1)
    maxFreq1 = np.max(meas_freq1)
    minFreq2 = np.min(meas_freq2)
    maxFreq2 = np.max(meas_freq2)

    minAmpl1 = np.min(meas_ampl1)
    maxAmpl1 = np.max(meas_ampl1)
    minAmpl2 = np.min(meas_ampl2)
    maxAmpl2 = np.max(meas_ampl2)

    minDur1 = np.min(meas_dur1)
    maxDur1 = np.max(meas_dur1)
    minDur2 = np.min(meas_dur2)
    maxDur2 = np.max(meas_dur2)
    
    bounds = np.tile(bound,(6,1)) #x y and z for both qubits
    
    optimizer = CMA(mean=mean, sigma=sigma, population_size=popsize, bounds=bounds)

    deltaFreq1 = maxFreq1 - minFreq1
    deltaAmpl1 = maxAmpl1 - minAmpl1
    deltaDur1 = maxDur1 - minDur1

    deltaFreq2 = maxFreq2 - minFreq2
    deltaAmpl2 = maxAmpl2 - minAmpl2
    deltaDur2 = maxDur2 - minDur2

    # coordinate = np.zeros((Ngen, 3))
    fidelity_mean = np.zeros((Ngen,2))

    for generation in range(Ngen):  
        normalized_mean = optimizer.mean
    
        mean = np.copy(normalized_mean)
        
        mean[0] = mean[0] * deltaFreq1 + minFreq1
        mean[1] = mean[1] * deltaAmpl1 + minAmpl1
        mean[2] = mean[2] * deltaDur1 + minDur1
        mean[3] = mean[3] * deltaFreq2 + minFreq2
        mean[4] = mean[4] * deltaAmpl2 + minAmpl2
        mean[5] = mean[5] * deltaDur2 + minDur2

        F = np.array([fidelityEstimate3D(mean[:3], meas_freq1, meas_ampl1, meas_dur1, meas_fid1)[0],
                      fidelityEstimate3D(mean[-3:], meas_freq2, meas_ampl2, meas_dur2, meas_fid2)[0]])
    
        fidelity_mean[generation] = F
    
        normalized_freq1 = np.zeros(popsize)
        normalized_ampl1 = np.zeros(popsize)
        normalized_dur1 = np.zeros(popsize)

        normalized_freq2 = np.zeros(popsize)
        normalized_ampl2 = np.zeros(popsize)
        normalized_dur2 = np.zeros(popsize)
        
        frequencies1 = np.zeros(popsize)
        amplitudes1 = np.zeros(popsize)
        durations1 = np.zeros(popsize)

        frequencies2 = np.zeros(popsize)
        amplitudes2 = np.zeros(popsize)
        durations2 = np.zeros(popsize)
        
        for i in range(popsize):
            x = optimizer.ask()
            
            normalized_freq1[i] = x[0]
            normalized_ampl1[i] = x[1]
            normalized_dur1[i] = x[2]
            
            normalized_freq2[i] = x[3]
            normalized_ampl2[i] = x[4]
            normalized_dur2[i] = x[5]

            frequencies1[i] = x[0]*deltaFreq1 + minFreq1
            amplitudes1[i] = x[1]*deltaAmpl1 + minAmpl1
            durations1[i] = x[2]*deltaDur1 + minDur1

            frequencies2[i] = x[3]*deltaFreq2 + minFreq2
            amplitudes2[i] = x[4]*deltaAmpl2 + minAmpl2
            durations2[i] = x[5]*deltaDur2 + minDur2
                    
        normalized_freqs_ampls_and_durs = [(xi, yi, zi, ai, bi, ci) for xi, yi, zi, ai, bi, ci in zip(normalized_freq1, normalized_ampl1, normalized_dur1, normalized_freq2, normalized_ampl2, normalized_dur2)]
        freqs_ampls_and_durs = np.array([np.array([xi, yi, zi, ai, bi, ci]) for xi, yi, zi, ai, bi, ci in zip(frequencies1, amplitudes1, durations1, frequencies2, amplitudes2, durations2)])
        objective_value = objective_2qb(freqs_ampls_and_durs, meas_freq1, meas_ampl1, meas_dur1, meas_fid1, meas_freq2, meas_ampl2, meas_dur2, meas_fid2)

        solutions = [(wi, vi) for wi, vi in zip(normalized_freqs_ampls_and_durs, objective_value)]
        optimizer.tell(solutions)

    # fidelity = fidelityEstimate3D(mean, meas_freq, meas_ampl, meas_dur, meas_fid)[0]
    return fidelity_mean
```

```python
#Import the data from measurements
duration = dataset.x0.data
frequency = dataset.x1.data
amplitude = dataset.x2.data
average_fidelity = dataset.y0.data
QND_pi_fidelity = dataset.y5.data

popsize = 10
sigma = 0.25 #1/4 of the domain 
mean = np.zeros(6,dtype=float)
Ngen = 50
Nreal = 1


fidelity_evolution1 = np.zeros((Nreal,Ngen))
fidelity_evolution2 = np.zeros((Nreal,Ngen))

for i in range(Nreal):
    fidelities_tmp = cmaes_frequency_amplitude_and_duration_2qb(mean, sigma, popsize, Ngen, 
                                                                frequency, amplitude, duration, average_fidelity,
                                                                frequency, amplitude, duration, average_fidelity)

    fidelity_evolution1[i,:] = fidelities_tmp[:,0]
    fidelity_evolution2[i,:] = fidelities_tmp[:,1]

        
mean_fidelity_evolution1 = np.mean(fidelity_evolution1,axis=0)
std_fidelity_evolution1 = np.std(fidelity_evolution1,axis=0)

mean_fidelity_evolution2 = np.mean(fidelity_evolution2,axis=0)
std_fidelity_evolution2 = np.std(fidelity_evolution2,axis=0)
```

```python
def objective_2qb(points, frequency1, amplitude1, duration1, fidelity1, frequency2, amplitude2, duration2, fidelity2):
    objective = np.zeros(len(points))
    for i in range(len(points)):
        F1 = fidelityEstimate3D(points[i][:3], frequency1, amplitude1, duration1, fidelity1)
        F2 = fidelityEstimate3D(points[i][-3:], frequency2, amplitude2, duration2, fidelity2) 
        objective[i] = np.log(1-F1)[0] + np.log(1-F2)[0]
    return objective
    
def cmaes_frequency_amplitude_and_duration_2qb_with_pop(mean, sigma, popsize, Ngen, 
                                           meas_freq1, meas_ampl1, meas_dur1, meas_fid1, 
                                           meas_freq2, meas_ampl2, meas_dur2, meas_fid2):
    bound = [0,1]
    
    minFreq1 = np.min(meas_freq1)
    maxFreq1 = np.max(meas_freq1)
    minFreq2 = np.min(meas_freq2)
    maxFreq2 = np.max(meas_freq2)

    minAmpl1 = np.min(meas_ampl1)
    maxAmpl1 = np.max(meas_ampl1)
    minAmpl2 = np.min(meas_ampl2)
    maxAmpl2 = np.max(meas_ampl2)

    minDur1 = np.min(meas_dur1)
    maxDur1 = np.max(meas_dur1)
    minDur2 = np.min(meas_dur2)
    maxDur2 = np.max(meas_dur2)
    
    bounds = np.tile(bound,(6,1)) #x y and z for both qubits
    
    optimizer = CMA(mean=mean, sigma=sigma, population_size=popsize, bounds=bounds)

    deltaFreq1 = maxFreq1 - minFreq1
    deltaAmpl1 = maxAmpl1 - minAmpl1
    deltaDur1 = maxDur1 - minDur1

    deltaFreq2 = maxFreq2 - minFreq2
    deltaAmpl2 = maxAmpl2 - minAmpl2
    deltaDur2 = maxDur2 - minDur2

    # coordinate = np.zeros((Ngen, 3))
    fidelity_mean = np.zeros((Ngen,2))
    fidelity_pop_q1 = np.zeros((Ngen,popsize))
    fidelity_pop_q2 = np.zeros((Ngen,popsize))

    track_freq1 = np.zeros(Ngen*popsize)
    track_ampl1 = np.zeros(Ngen*popsize)

    for generation in range(Ngen):  
        normalized_mean = optimizer.mean
    
        mean = np.copy(normalized_mean)
        
        mean[0] = mean[0] * deltaFreq1 + minFreq1
        mean[1] = mean[1] * deltaAmpl1 + minAmpl1
        mean[2] = mean[2] * deltaDur1 + minDur1
        mean[3] = mean[3] * deltaFreq2 + minFreq2
        mean[4] = mean[4] * deltaAmpl2 + minAmpl2
        mean[5] = mean[5] * deltaDur2 + minDur2

        F = np.array([fidelityEstimate3D(mean[:3], meas_freq1, meas_ampl1, meas_dur1, meas_fid1)[0],
                      fidelityEstimate3D(mean[-3:], meas_freq2, meas_ampl2, meas_dur2, meas_fid2)[0]])
    
        fidelity_mean[generation] = F
    
        normalized_freq1 = np.zeros(popsize)
        normalized_ampl1 = np.zeros(popsize)
        normalized_dur1 = np.zeros(popsize)

        normalized_freq2 = np.zeros(popsize)
        normalized_ampl2 = np.zeros(popsize)
        normalized_dur2 = np.zeros(popsize)
        
        frequencies1 = np.zeros(popsize)
        amplitudes1 = np.zeros(popsize)
        durations1 = np.zeros(popsize)

        frequencies2 = np.zeros(popsize)
        amplitudes2 = np.zeros(popsize)
        durations2 = np.zeros(popsize)
        
        for i in range(popsize):
            x = optimizer.ask()
            
            normalized_freq1[i] = x[0]
            normalized_ampl1[i] = x[1]
            normalized_dur1[i] = x[2]
            
            normalized_freq2[i] = x[3]
            normalized_ampl2[i] = x[4]
            normalized_dur2[i] = x[5]

            frequencies1[i] = x[0]*deltaFreq1 + minFreq1
            amplitudes1[i] = x[1]*deltaAmpl1 + minAmpl1
            durations1[i] = x[2]*deltaDur1 + minDur1

            frequencies2[i] = x[3]*deltaFreq2 + minFreq2
            amplitudes2[i] = x[4]*deltaAmpl2 + minAmpl2
            durations2[i] = x[5]*deltaDur2 + minDur2

            fidelity_pop_q1[generation][i] = fidelityEstimate3D(np.array([frequencies1[i], amplitudes1[i], durations1[1]]),meas_freq1,meas_ampl1,meas_dur1,meas_fid1)[0]
            fidelity_pop_q2[generation][i] = fidelityEstimate3D(np.array([frequencies2[i], amplitudes2[i], durations2[1]]),meas_freq2,meas_ampl2,meas_dur2,meas_fid2)[0]

            track_freq1[popsize*generation + i] = frequencies1[i]
            track_ampl1[popsize*generation + i] = amplitudes1[i]

        
        normalized_freqs_ampls_and_durs = [(xi, yi, zi, ai, bi, ci) for xi, yi, zi, ai, bi, ci in zip(normalized_freq1, normalized_ampl1, normalized_dur1, normalized_freq2, normalized_ampl2, normalized_dur2)]
        freqs_ampls_and_durs = np.array([np.array([xi, yi, zi, ai, bi, ci]) for xi, yi, zi, ai, bi, ci in zip(frequencies1, amplitudes1, durations1, frequencies2, amplitudes2, durations2)])
        objective_value = objective_2qb(freqs_ampls_and_durs, meas_freq1, meas_ampl1, meas_dur1, meas_fid1, meas_freq2, meas_ampl2, meas_dur2, meas_fid2)

        solutions = [(wi, vi) for wi, vi in zip(normalized_freqs_ampls_and_durs, objective_value)]
        optimizer.tell(solutions)

    # fidelity = fidelityEstimate3D(mean, meas_freq, meas_ampl, meas_dur, meas_fid)[0]
    return fidelity_mean, fidelity_pop_q1, fidelity_pop_q2, track_freq1, track_ampl1
```
