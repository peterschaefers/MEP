```python
from quantify_core.data.handling import (
    default_datadir,
    get_latest_tuid,
    get_tuids_containing,
    load_dataset,
    locate_experiment_container,
    set_datadir,
)

import numpy as np
import matplotlib.pyplot as plt
```

```python
set_datadir(default_datadir())
```

```python
tuids = get_tuids_containing(contains="20250402")

datasets = {}  # Dictionary to store datasets

for i, tuid in enumerate(tuids):
    datasets[f"dataset{i}"] = load_dataset(tuid)
```

```python
popsize = 16

total = int(len(datasets["dataset0"]["x0"]))
mean_indices = np.arange(popsize, total, popsize + 1)

labels = range(popsize)
# print(mean_indices)

for key in datasets:
    data = datasets[key]  # Retrieve the dataset
    sc = plt.scatter(data["x0"], data["x1"], c=data["x2"], cmap="viridis")  # Use x2 for color
    plt.colorbar(sc, label="Duration")  # Add a colorbar
    plt.xlabel("Frequency")
    plt.ylabel("Amplitude")
    plt.title(f"Scatter plot of the mean for {key}")
    plt.show()
```

```python
popsize = 16

total = int(len(datasets["dataset0"]["x0"]))
mean_indices = np.arange(popsize, total, popsize + 1)

labels = range(1, 1+ len(mean_indices))  # Adjusted to match number of labeled points

for key in datasets:
    data = datasets[key]  # Retrieve the dataset
    x_vals = data["x0"][mean_indices]
    y_vals = data["x1"][mean_indices]
    colors = data["x2"][mean_indices]

    sc = plt.scatter(x_vals, y_vals, c=colors, cmap="viridis")  # Use x2 for color

    # Add labels
    for i, (x, y) in enumerate(zip(x_vals, y_vals)):
        plt.annotate(f'{labels[i]}', (x, y), textcoords="offset points", xytext=(5,5), ha='center', fontsize=5, color='black')

    plt.colorbar(sc, label="Duration")  # Add a colorbar
    
    plt.xlabel("Frequency")
    plt.ylabel("Amplitude")
    plt.xlim(7.55*1e9,7.65*1e9)
    plt.ylim(0.025,0.070)
    plt.title(f"Scatter plot of the mean for {key}")
    plt.show()
```
