```python
from pyDOE2 import lhs
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
# # Generate Latin Hypercube samples
samples = lhs(n=3, samples=6, criterion='maximin')
print(samples.T)
print(samples[0,1])

# Split into coordinates
x = samples[:, 0]
y = samples[:, 1]
z = samples[:, 2]

# All 2D projections
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].scatter(x, y)
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_title('X vs Y')

axes[1].scatter(x, z)
axes[1].set_xlabel('X')
axes[1].set_ylabel('Z')
axes[1].set_title('X vs Z')

axes[2].scatter(y, z)
axes[2].set_xlabel('Y')
axes[2].set_ylabel('Z')
axes[2].set_title('Y vs Z')

plt.tight_layout()
plt.show()

print(samples)
```
